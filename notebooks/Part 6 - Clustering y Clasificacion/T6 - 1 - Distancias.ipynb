# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../../datasets/movies/movies.csv", sep=";")
data.head()
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1)
dd2 = distance_matrix(data[movies], data[movies], p=2)
dd10 = distance_matrix(data[movies], data[movies], p=10)
```

```python
def dm_to_df(dd, col_name):
    import pandas as pd
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
dm_to_df(dd1, data["user_id"])
```

```python
dm_to_df(dd2, data["user_id"])
```

```python
dm_to_df(dd10, data["user_id"])
```

# Clustering jerrquico

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], method="ward")
Z
```

```python
plt.figure(figsize=(25, 10))
plt.title("Dendrograma jerrquico para el clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90.0, leaf_font_size=10.0)
plt.show()
```

```python
plt.figure(figsize=(25, 10))
Z = linkage(data[movies], method="average")
plt.title("Dendrograma jerrquico para el clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90.0, leaf_font_size=10.0)
plt.show()
```

```python
data[movies]
```

```python
plt.figure(figsize=(25, 10))
Z = linkage(data[movies], method="complete")
plt.title("Dendrograma jerrquico para el clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90.0, leaf_font_size=10.0)
plt.show()
```

```python
plt.figure(figsize=(25, 10))
Z = linkage(data[movies], method="single", metric="cosine")
plt.title("Dendrograma jerrquico para el clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90.0, leaf_font_size=10.0)
plt.show()
```
